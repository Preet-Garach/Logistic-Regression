```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
```

```python
data = pd.read_csv('bank-full.csv', sep=';')
data.head()
```

```python
data.info()
```

## One Hot Encoding

```python
data1 = pd.get_dummies(data, columns= ['job','marital','education','contact','poutcome','month'])
data1
```

```python
pd.set_option('display.max_columns',None)
data1
```

```python
data1.info()
```

## Binary Encoding

```python
data1['default'] = np.where(data1['default'].str.contains("yes"), 1, 0)
data1['loan'] = np.where(data1['loan'].str.contains("yes"), 1, 0)
data1['housing'] = np.where(data1['housing'].str.contains("yes"), 1, 0)
data1['y'] = np.where(data1['y'].str.contains("yes"), 1, 0)
data1
```

## Standardization of 'age' 

```python
from sklearn.preprocessing import StandardScaler
data1[['age']] = StandardScaler().fit_transform(data1[['age']])
data1.head()
```

## Model Building

```python
X = pd.concat([data1.iloc[:,0:10],data1.iloc[:,11:]],axis=1)
Y = data1.iloc[:,10]
```

```python
classifier = LogisticRegression()
classifier.fit(X,Y)
```

## Model Prediction

```python
y_pred=classifier.predict(X)
y_pred
```

```python
y_pred_df=pd.DataFrame({'actual_y':Y,'y_pred':y_pred})
y_pred_df
```

## Testing Model Accuracy

```python
confusion_matrix = confusion_matrix(Y,y_pred)
confusion_matrix
```

```python
classifier.predict_proba(X)[:,1]
```

```python
fpr,tpr,thresholds = roc_curve(Y,classifier.predict_proba(X)[:,1])
plt.plot(fpr,tpr,color='red')
auc = roc_auc_score(Y,y_pred)

plt.plot(fpr,tpr,color='red',label='logit model(area  = %0.2f)'%auc)
plt.plot([0,1],[0,1],'k--')
plt.xlabel('False Positive Rate or [1 - True Negative Rate]')
plt.ylabel('True Positive Rate')
plt.show()

print('auc accuracy:',auc)
```
